```python
url = "/kaggle/input/datasets/durdanajenisbaeva/tashkent-house-data/uybor.xlsx"
df = pd.read_excel(url)
df
```

```python
df_avg_price = df.groupby('district')['price'].mean()
df_avg_price = df_avg_price.astype(int)
sns.barplot(x=df_avg_price.index, y=df_avg_price.values)
plt.xticks(rotation=90)
plt.show()
```

```python
sns.displot(data=df, x='price', hue= 'district')
plt.show()
```

```python
sns.scatterplot(data=df, x='size', y='price')
plt.show()
```

```python
apart_count = df.groupby('district')['address'].count()
sns.barplot(x=apart_count.index, y=apart_count.values)
plt.xticks(rotation=90)
plt.show()
```

```python
room_price = df.groupby('rooms')['price'].mean()
room_price = room_price.astype(int)
sns.regplot(x=room_price.index, y=room_price.values)
plt.show()
```
